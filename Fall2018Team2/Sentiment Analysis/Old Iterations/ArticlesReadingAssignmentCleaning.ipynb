# Articles Reading Assignment Cleaning

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("Classified Articles.csv")

df2 = pd.read_csv("Articles Reading Assignment.csv")
df2 = df2.dropna()
df2["Sentiment"] += 1
df2["Sentiment"] /= 2

df2["Content"] = ["" for i in range(len(df2))]
df2["Content Length"] = [0 for i in range(len(df2))]

for i, row in df2.iterrows():
    x = row["URL"]
    
    key_words = df1[df1["source_url"] == x][:1]["contents"].values[0]
    df2.at[i, "Content"] = str(key_words)
    df2.at[i, "Content Length"] = len(key_words)
```
